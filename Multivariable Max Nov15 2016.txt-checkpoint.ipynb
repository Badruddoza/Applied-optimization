$\text{Economically optimum input use-levels}$

A profit function is given where $P_y,P_1,$ and $P_2$ are the prices received and paid for the output and inputs, respectively, and FC represents fixed costs.

(2.30a)$$\pi=P_yY-P_1X_1-P_2X_2-FC$$

(2.30b)$$\pi=P_yh(X_1,X_2|X_3,X_4)-P_1X_1-P_2X_2-FC$$

Notice that the cost function is $C=P_1X_1+P_2X_2+FC$. Substituting the production function (2.23) for $Y$ in (2.30), with both $X_1$ and $X_2$ variable, the profit equation becomes (2.30b). The assumed goal of the producer is profit maximization. Setting the first-order partial derivatives of (2.30b) with respect to $X_1$ and $X_2$ equal to zero, a system of two equations is obtained:

(2.31)$$\frac{\partial \pi}{\partial X_1}=P_y \left( \frac{\partial Y}{\partial X_1} \right)-P_1=0\Rightarrow \frac{\partial Y}{\partial X_1}=\frac{P_1}{P_y}$$

$$ \frac{\partial \pi}{\partial X_2}=P_y \left( \frac{\partial Y}{\partial X_2} \right)-P_2=0 \Rightarrow \frac{\partial Y}{\partial X_2}=\frac{P_2}{P_y}$$

Solving this system simultaneously, we have the quantity each of $X_1$ and $X_2$ that will maximize profits relative to fixed resources.

Intuitively, the above equations mean that the value of marginal product of each input should equate their per unit cost of usage. That is, marginal revenue equates marginal cost at optimal choice of inputs. By 'optimal' we mean profit maximizing.

## Employ the model in real world problem

$\text{Problem 1: 2 inputs 1 output}$

Consider a profit function for a farmer who uses water $X_1$ and fertilizer $X_2$ to produce a crop $Y$. The prices are $P_1, P_2,$ and $P_y$, respectively. Assume there is no fixed cost, $FC=0$. No factor of production can be negative. The farmer solves:

$$\max_{X_1,X_2,L} \pi=P_yY-P_1X_1-P_2X_2$$

$$\Rightarrow \max_{X_1,X_2} \pi=P_y(KX_1^{b_1}X_2^{b_2})-P_1X_1-P_2X_2$$

$$\text{s.t.  } X_1\geq0, X_2\geq0$$

To find a solution using Python, use the following parameter values:

$$P_y=20, P_1=1.35, P_2=1.50$$

$$k=1, b_1=0.2, b_2=0.3$$

Find the optimal uses of water and fertilizer to produce the crop at optimal profit.

```python
""" Create environment """
import numpy as np
from scipy.optimize import minimize
py=20.0
p1=1.35
p2=1.5
b1=0.2
b2=0.3
k=1

""" Define profit function """
""" (we use negative sign before minimize command to maximize our function) """
def func(x, sign=-1.0):
    return sign*(py*k*(x[0]**b1)*(x[1]**b2)-p1*x[0]-p2*x[1])
""" Derivatives with respect to water and fertilizer """
def func_deriv(x, sign=-1.0):
    dfdx0 = sign*(py*k*b1*(x[0]**(b1-1))*(x[1]**b2)-p1)
    dfdx1 = sign*(py*k*(x[0]**b1)*b2*(x[1]**(b2-1))-p2)
    return np.array([ dfdx0, dfdx1])
""" Capacity constraints and their derivatives """
cons = ({'type': 'ineq',
         'fun' : lambda x: np.array([x[0]]),
         'jac' : lambda x: np.array([1.0, 0.0])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([x[1]]),
         'jac' : lambda x: np.array([0.0, 1.0])})
"""Minimize the function with inequality constraints using SLSQP method"""
res = minimize(func, [1.0,1.0], args=(-1.0,), jac=func_deriv, 
               constraints=cons, method='SLSQP', options={'disp': True})
print("optimal water (X1) and fertilizer (X2)",np.around(res.x,decimals=2))
```

$\text{Problem 2: 1 input 2 outputs}$

Now assume the farmer produces two crops $Y_1$ and $Y_2$. The farmer uses water $X$ only and the prices are $P_1, P_2, \text{ and } w_1$, respectively. $x$ cannot be negative.

Production function for crop $Y_1$ and $Y_2$ are, respectively: 

$$Y_1=X_1^{a}$$

$$Y_2=X_1^{b}$$

Therefore the farmer solves the following problem:

$$\max_{X} \pi=P_1Y_1+P_2Y_2-wX$$

$$\Rightarrow \max_{X} \pi=P_1X^{a}+P_2X^{b}-wX$$

$$\text{s.t.  } X\geq0$$

To find a solution using Python, use the following parameter values:

$$P_1=20, P_2=30, w=1.35$$

$$a=0.2, b=0.3$$

Find the uses of water and optimal output of each crop that maximize profit.

```python
""" Create environment """
import numpy as np
from scipy.optimize import minimize_scalar
""" Set parameters """
p1=20.0
p2=30.0
w=1.35
a=0.2
b=0.3
infinity=1e99
""" if we minimize negative of our function then it is equivalent to the maximize command """
f = lambda x: -(p1*(x**a)+p2*(x**b)-w*x)
res = minimize_scalar(f, bounds=(0,infinity), method='bounded')
print('water applied=',np.around(res.x,decimals=2))
print('Y1=',np.around(res.x**a,decimals=2))
print('Y2=',np.around(res.x**b,decimals=2))
print('maximum profit=', np.around(p1*(res.x**a)+p2*(res.x**b)-w*res.x,decimals=2))
```

$\text{Problem 3: 2 inputs 2 outputs}$

We now proceed to more realistic scenario of multiple inputs and outputs. The farmer produces two crops $Y_1$ and $Y_2$. The farmer uses water $X_1$ and fertilizer $X_2$ and the prices are $P_1, P_2, w_1$ and $w_2$, respectively. The cost function has additive form $C=w_1X_1+w_2X_2$. Water and fertilizer take only non-negative values.

Production function for crop $Y_1$ and $Y_2$ are, respectively: 

$$Y_1=X_1^{a_1}X_2^{a_2}$$

$$Y_2=X_1^{b_1}X_2^{b_2}$$

Therefore the farmer solves the following problem:

$$\max_{X_1,X_2} \pi=P_1Y_1+P_2Y_2-w_1X_1-w_2X_2$$

$$\Rightarrow \max_{X_1,X_2} \pi=P_1(X_1^{a_1}X_2^{a_2})+P_2(X_1^{b_1}X_2^{b_2})-w_1X_1-w_2X_2$$

$$\text{s.t.  } X_1\geq0, X_2\geq0$$

To find a solution using Python, use the following parameter values:

$$P_1=2.0, P_2=3.0, w_1=1.35, w_2=1.50$$

The production elasticities of water $X_1$ and fertilizer $X_2$ are:

$$\begin{bmatrix} a_1 & a_2\\ b_1 & b_2 \end{bmatrix}=\begin{bmatrix} 0.2 & 0.3\\ 0.4 & 0.5 \end{bmatrix}$$

Intuitively, a percent increase in $X_1$ increases production of $Y_1$ by $a_1=0.2$ percent and so on.

Find the optimal uses of water and fertilizer to produce the crop at optimal profit.

```python
""" Create environment """
import numpy as np
from scipy.optimize import minimize
p1=2.0
p2=3.0
w1=1.35
w2=1.50
a1=0.2
a2=0.3
b1=0.4
b2=0.5

""" Define profit function """
""" (we use negative sign before minimize command to maximize our function) """
def func(x, sign=-1.0):
    return sign*(p1*(x[0]**a1)*(x[1]**a2)+p2*(x[0]**b1)*(x[1]**b2)-w1*x[0]-w2*x[1])
""" Derivatives with respect to water and fertilizer """
def func_deriv(x, sign=-1.0):
    dfdx0 = sign*(a1*p1*x[0]**(a1-1)*x[1]**a2+b1*p2*x[0]**(b1-1)*x[1]**b2-w1)
    dfdx1 = sign*(a2*p1*x[0]**a1*x[1]**(a2-1)+b2*p2*x[0]**b1*x[1]**(b2-1)-w2)
    return np.array([ dfdx0, dfdx1])
""" Capacity constraints and their derivatives """
cons = ({'type': 'ineq',
         'fun' : lambda x: np.array([x[0]]),
         'jac' : lambda x: np.array([1.0, 0.0])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([x[1]]),
         'jac' : lambda x: np.array([0.0, 1.0])})
"""Minimize the function with inequality constraints using SLSQP method"""
res = minimize(func, [1.0,1.0], args=(-1.0,), jac=func_deriv, 
               constraints=cons, method='SLSQP', options={'disp': True})

print("optimal water (X1)=",np.around(res.x[0],decimals=2))
print("optimal fertilizer (X2)=",np.around(res.x[1],decimals=2))
print('one crop (Y1)=',np.around(res.x[0]**a1*res.x[1]**a2,decimals=2))
print('another crop (Y2)=',np.around(res.x[0]**b1*res.x[1]**b2,decimals=2))
print('maximum profit=', np.around( p1*(res.x[0]**a1)*(res.x[1]**a2)+p2*(res.x[0]**b1)
                                   *(res.x[1]**b2)-w1*res.x[0]-w2*res.x[1] ,decimals=2) )
```

$\text{Problem 4: 2 inputs 2 outputs (introducing land constraint)}$

So far we considered land unlimited. The following problem assumes the same specifications as in Problem(3), but it includes land as factor of production. Instead of fertilizer, let $X_2$ be the amount of land used for production. Total land available to the farmer is $\bar{L}$. Notice that if $X_2=0$ then no production is possible. So we must set $0 < X_2 \leq \bar{L}$

Production function for crop $Y_1$ and $Y_2$ are, respectively: 

$$Y_1=X_1^{a_1}X_2^{a_2}$$

$$Y_2=X_1^{b_1}X_2^{b_2}$$

Therefore the farmer solves the following problem:

$$\max_{X_1,X_2} \pi=P_1Y_1+P_2Y_2-w_1X_1-w_2X_2$$

$$\Rightarrow \max_{X_1,X_2} \pi=P_1(X_1^{a_1}X_2^{a_2})+P_2(X_1^{b_1}X_2^{b_2})-w_1X_1-w_2X_2$$

$$\text{s.t.  } X_1\geq0$$ 

$$0< X_2\leq \bar{L}\Rightarrow \bar{L}-X_2 \geq0$$

To find a solution using Python, use the same parameter values:

$$P_1=2.0, P_2=3.0, w_1=1.35, w_2=1.50, \bar{L}=4$$

Let the production elasticities of water $X_1$ and land $X_2$ be:

$$\begin{bmatrix} a_1 & a_2\\ b_1 & b_2 \end{bmatrix}=\begin{bmatrix} 0.2 & 0.3\\ 0.4 & 0.5 \end{bmatrix}$$

Intuitively, a percent increase in $X_1$ increases production of $Y_1$ by $a_1=0.2$ percent and so on.

Find the optimal uses of water and fertilizer to produce the crop at optimal profit.

```python
""" Create environment """
import numpy as np
from scipy.optimize import minimize
p1=2.0
p2=3.0
w1=1.35
w2=1.50
a1=0.2
a2=0.3
b1=0.4
b2=0.5
lbar=4.0

""" Define profit function """
""" (we use negative sign before minimize command to maximize our function) """
def func(x, sign=-1.0):
    return sign*(p1*(x[0]**a1)*(x[1]**a2)+p2*(x[0]**b1)*(x[1]**b2)-w1*x[0]-w2*x[1])
""" Derivatives with respect to water and fertilizer """
def func_deriv(x, sign=-1.0):
    dfdx0 = sign*(a1*p1*x[0]**(a1-1)*x[1]**a2+b1*p2*x[0]**(b1-1)*x[1]**b2-w1)
    dfdx1 = sign*(a2*p1*x[0]**a1*x[1]**(a2-1)+b2*p2*x[0]**b1*x[1]**(b2-1)-w2)
    return np.array([ dfdx0, dfdx1])
""" Capacity constraints and their derivatives """
cons = ({'type': 'ineq',
         'fun' : lambda x: np.array([x[0]]),
         'jac' : lambda x: np.array([1.0, 0.0])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([lbar-x[1]]),
         'jac' : lambda x: np.array([0.0, -1.0])})
"""Minimize the function with inequality constraints using SLSQP method"""
res = minimize(func, [1.0,1.0], args=(-1.0,), jac=func_deriv, 
               constraints=cons, method='SLSQP', options={'disp': True})

print("optimal water (X1)=",np.around(res.x[0],decimals=2))
print("optimal land (X2)=",np.around(res.x[1],decimals=2))
print('one crop (Y1)=',np.around(res.x[0]**a1*res.x[1]**a2,decimals=2))
print('another crop (Y2)=',np.around(res.x[0]**b1*res.x[1]**b2,decimals=2))
print('maximum profit=', np.around( p1*(res.x[0]**a1)*(res.x[1]**a2)+p2*(res.x[0]**b1)
                                   *(res.x[1]**b2)-w1*res.x[0]-w2*res.x[1] ,decimals=2) )
```

$\text{Problem 5: General model with 3 factors of production (limited land) and 2 outputs}$

Now we are ready to proceed towards more general problems. With all assumptions in Problem(4), consider another input $X_2$ (say, fertilizer). For simplicity, assume that $X_2$ is required for crop one ($Y_1$) only. Use the following parameters to find the optimal X's and Y's in Python.

Production function of $Y_1$ is

$$Y_1=X_1^{a_1}X_2^{a_2}X_3^{a_3}$$

where production elasticity of fertilizer is, $a_3=0.4$

Price of each unit of fertilizer is, $w_3=2.0$

Hence the farmer solves the following problem

$$\Rightarrow \max_{X_1,X_2} \pi=P_1(X_1^{a_1}X_2^{a_2}X_3^{a_3})+P_2(X_1^{b_1}X_2^{b_2})-w_1X_1-w_2X_2-w_3X_3$$

$$\text{s.t.  } X_1\geq0$$ 

$$0\leq X_2\leq \bar{L}\Rightarrow \bar{L}-X_2 \geq0$$

$$X_3\geq0$$

```python
""" Create environment """
import numpy as np
from scipy.optimize import minimize
p1=2.0
p2=3.0
w1=1.35
w2=1.50
w3=2.0
a1=0.2
a2=0.3
a3=0.4
b1=0.4
b2=0.5
lbar=4.0

""" Define profit function """
""" (we use negative sign before minimize command to maximize our function) """
def func(x, sign=-1.0):
    return sign*(p1*(x[0]**a1)*(x[1]**a2)*(x[2]**a3)+p2*(x[0]**b1)*(x[1]**b2)-w1*x[0]-w2*x[1]-w3*x[2])
""" Derivatives with respect to water and fertilizer """
def func_deriv(x, sign=-1.0):
    dfdx0 = sign*(a1*p1*x[0]**(a1-1)*x[1]**a2*x[2]**a3+b1*p2*x[0]**(b1-1)*x[1]**b2-w1)
    dfdx1 = sign*(a2*p1*x[0]**a1*x[1]**(a2-1)*x[2]**a3+b2*p2*x[0]**b1*x[1]**(b2-1)-w2)
    dfdx2 = sign*(a3*p1*x[0]**a1*x[1]**a2*x[2]**(a3-1)-w3)
    return np.array([ dfdx0, dfdx1, dfdx2])
""" Capacity constraints and their derivatives """
cons = ({'type': 'ineq',
         'fun' : lambda x: np.array([x[0]]),
         'jac' : lambda x: np.array([1.0, 0.0, 0.0])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([lbar-x[1]]),
         'jac' : lambda x: np.array([0.0, -1.0, 0.0])},
        {'type': 'ineq',
        'fun' : lambda x: np.array([x[2]]),
        'jac' : lambda x: np.array([0.0, 0.0, 1.0])})

"""Minimize the function with inequality constraints using SLSQP method"""
res = minimize(func, [1.0,1.0,1.0], args=(-1.0,), jac=func_deriv, 
               constraints=cons, method='SLSQP', options={'disp': True})

print("optimal water (X1)=",np.around(res.x[0],decimals=2))
print("optimal land (X2)=",np.around(res.x[1],decimals=2))
print("optimal fertilizer (X3)=",np.around(res.x[2],decimals=2))
print('one crop (Y1)=',np.around(res.x[0]**a1*res.x[1]**a2*res.x[2]**a3,decimals=2))
print('another crop (Y2)=',np.around(res.x[0]**b1*res.x[1]**b2,decimals=2))
print('maximum profit=', np.around( p1*(res.x[0]**a1)*(res.x[1]**a2)*(res.x[2]**a3)+p2*(res.x[0]**b1)
                                   *(res.x[1]**b2)-w1*res.x[0]-w2*res.x[1]-w3*res.x[2],decimals=2) )
```

$$\text{s.t.}\quad p_1x_1^2+p_2x_2=\bar{C}$$

Let $a=0.4, b=0.5, p_1=2, p_2=3$

```python
""" Create environment """
import numpy as np
from scipy.optimize import minimize
a=0.4
b=0.5
p1=2
p2=3
cbar=100

def func(x, sign=-1.0):
    """ negative sign for maximization """
    return sign*( (x[0]**a) * (x[1]**b))
def func_deriv(x, sign=-1.0):
    """ Derivative of objective function """
    dfdx0 = sign*(a*x[0]**(a-1)*x[1]**b)
    dfdx1 = sign*(x[0]**(a)*b*x[1]**(b-1))
    return np.array([ dfdx0, dfdx1 ])
cons = ({'type': 'eq',
         'fun' : lambda x: np.array([p1*x[0]**2 + p2*x[1]-cbar]),
         'jac' : lambda x: np.array([2*p1*x[0], p2])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([x[1]]),
         'jac' : lambda x: np.array([0.0, 1.0])})
res = minimize(func, [1.0,1.0], args=(-1.0,), jac=func_deriv,
               constraints=cons, method='SLSQP', options={'disp': True})
print("x_1 and x_2", res.x)
```

## Literature on Curve Fitting

In applied Economic papers, regression techniques are among some of the widely used methods to fit a curve. Many studies utilize appropriate polynomial regression to capture the curvature (for example, see Hexem and Heady, 1978). Several recent studies like Schlenker and Roberts (2009) mostly focus on three methods to compare the results: (1) a step function that fits separate segments of particular range in the domain, (2) higher order (Chebychev's) polynomial function of temperature and 95\% confidence band, sometimes after adjusting for spatial correlation, and (3) piecewise linear specification. In practice, different variety of each approach exists. For example, Levy (1959) derives the model from the ratio of two series of polynomials. Nonparametric techniques, on the contrary, contain smoothing splines and kernel smoothers with variable bandwidth (Muller and Stadmuller, 1987; Fan Gijbels, 1995). Hastie and Tibshirani (1990) discuss generalized additive models of both genres. Silverman (1985) designs a model with roughness penalty.\\

However, each method has its advantages and disadvantages. For instance, polynomial regression suffers from the fact that individual observations can have an influence, in unexpected ways, on remote parts of the curve. Moreover, the global nature of polynomials might insufficiently estimate wiggly curves (Denison et al., 1998). The cubic spline occasionally lacks the production of extraneous inflection points in the curve (Cline, 1974). To capture the numerous extraneous influences, a 'free hand' drawing can be the most convenient but generally subjective in nature. As Akima (1970) points out, we do not assume any functional form for the whole curve when we try to fit a smooth curve manually; rather we draw a portion of the curve based on local aspect. This local aspect is difficult to adopt in a mathematical model that is differentiable. Put differently, a major limitation of piecewise approach is the continuity of first-order derivative at the junction points cannot be guaranteed.\\ 

A general approach to curve fitting, therefore, is to allow the function to be a piecewise polynomial made up of low order pieces only between adjacent knot points. The osculatory interpolation method is one that depends on piecewise function with third degree polynomials with slopes at the junction points that are locally determined (Ackland, 1915; Karup, 1899). Albeit such method determines the slope of the curve at each given point locally to form a smooth curve by peicewise application, it also gives unnatural curves in some cases (Akima, 1970). Hence, Akima (1969, 1970) proposes a method of interpolation and smooth curve fitting that is based on a piecewise function with slopes at the junction points locally determined under a geometrical condition.\\ 

Nevertheless, in step functions and piecewise linear specification, there is no definite rule of thumb for splitting the domain. This leads to some parametric examples that involve spline fitting with adaptive knot placement (see Smith, 1982; Friedman and Silverman, 1989; and Friedman, 1991). For example, Denison et al. (1998) proposes a Bayesian approach that models the function by a piecewise polynomial assuming the number of knots is unknown at unknown location. Between each knot they fit a fixed low order polynomial and the knots are treated as parameters to be inferred. Green (1995) studies Markov chain Monte Carlo simulation technique of reversible jumps to calculate of posterior distributions for number and location of knots. \\

## Estimating Water Response Function

We analyze lab-generated data of 21 crops from nine regions in Washington state. The purpose of the exercise is to estimate the marginal effect of irrigation on crop yield, thereby to figure out how does the response function look like. Irrigation holding other things constant has diminishing marginal productivity. Graphically, the crop yield 'decreasingly increases' in irrigation; conforming a shape that looks like an inverted 'u'. For this, we use a simple model discussed by Hexem and Heady (1978); the crop yield per acre $Y$ depends on acre-inches of water $X_1$, and some other variables like pounds per acre of fertilizer applied $X_2$, and plant population $X_3$. That is,

$$Y=f(X_1, X_2, X_3, ...)$$

and if $Y$ reaches its maximum, $Y=Y^*$ where $X_1={X_1}^*$

The marginal physical product of irrigation ($X_1$) is

$\dfrac{dY}{dX_1}=\dfrac{df(X_1|X_2, X_3, ...)}{dX_1}$

Since the actual relationship between $Y$ and $X_1$ is unknown, we proceed with cubic polynomial under the assumption that $f(x)$ is at least three times differentiable with respect to $X_1$. Such polynomial form comes from a concept known as Taylor's (1969) series expansion. Albeit a quadratic polynomial suffices for inverted 'u' shaped function, we included one higher order to capture the curvature at end points, if there is any. Including up to cubic terms of $X_1$ the model becomes

$$f(X)=a_0+a_1X_1+a_2{X_1}^2+a_3{X_1}^3+X_2+X_3+...$$

Two problems arise in econometric estimation of the model. First, the data set we use is lab-generated and it does not provide us with the points where irrigation is overused ($X_1>X_1^*$). The dependent variable ($Y$), therefore, ranges from zero to $Y^*$ only over the domain where $X_1\leq X_1^*$. To resolve the issue, we propose a 'mirror-domain' strategy (see Figure 1). We construct points of $Y$ at $X_1>X_1^*$ that are just mirror images of existing $Y$ where $X_1<{X_1}^*$. For instance, if $Y=2$ where $X_1=50$, and we know that the maximum, $Y^*=4$ takes place at $X_1^*=100$, we can generate $Y=2$ where $X_1=(100+50)=150$. Thus, we have a point of $Y$ where $X_1>X_1^*$. This assimilates the data to the inverted 'u'-type functional form, thereby allows us to estimate the model above.

Second, we do not have information about variables that might have crucial impacts on crop yield; such as soil quality, fertilizer used and plant population. However, it is less likely that the variables are correlated with irrigated amount of water under careful lab-experiment. This reduces the plausibility of omitted variable bias. We estimated the following model with a mirror domain of irrigation

$$Y=a_0+a_1X+a_2X^2+a_3X^3+Z_1+Z_2+\epsilon$$

Where, $Y=$crop yield in tons per acre, $X=$irrigation per foot, $Z_1$=precipitation, $Z_2$=year, and $\epsilon\sim N(0,\sigma^2)$. Not surprisingly, precipitation and rainfall have high and significant correlation (about 95 percent). Precipitation and rainfall both have significant and negative correlation with irrigation. The magnitude is, however, very low (less than 4 percent). Since precipitation and some other variables (such as rainfall and growing degree days) are highly correlated, and since precipitation alone explains the data better (evident form $F$ statistics and adjusted $R$), we keep precipitation only. The variable 'Year' captures changes in time. We did not include the regional dummy variables, because it is not certain that regions generate an intercept effect on the model. Instead, we separately ran the regressions for each (21) crop and (9) district. Not every crop was cultivated in each district. Therefore, our total number of regression equations is 104 rather than $(21*9=)$ 189.

Scrutinizing adjusted $R^2$ and fitted plots, we conclude that the cubic model and mirror-domain strategy reasonably fits the data. Mean adjusted $R^2=0.588$ with standard deviation of $0.125$. None of the 104 parameters adjacent to cubic term of irrigation ($a_3$ in the model above) turn out to be statistically (or economically) significant. However, all of the parameters with linear and quadratic terms of irrigation ($a_1$ and $a_2$) are highly significant. This corroborates our hypothesis that crop yield follows an inverted 'u'-shaped path as irrigation increases. The crop that exhibits discontinuous or step function type pattern gets relatively poorer fit (Corn in Figure 1 is an example). Apple in Benton and Kennewick districts is the least likely to be modeled by cubic function of irrigation (about 30 percent). The model performs the best in predicting sweet corn yields in Cascade, Kennewick, Roza, and Wapato project (over 80 percent) (see Figure 2). Grapejuice and grapes have plots that look much like horizontal lines, on the other hand, Potatoes, asparagus, and corn follow some step-like pattern.

## Two studies on agricultural input demand

Griliches, Zvi. "The demand for fertilizer: An economic interpretation of a technical change." Journal of Farm Economics 40, no. 3 (1958): 591-606.

 The paper explains the changes in fertilizer use in U.S. farms with the simple time series framework. Previously done studies did not find any statistically significant relationship between the price paid for fertilizer and fertilizer consumption. The contribution of the paper is, hence, using a model that takes 'real' values of fertilizer use (free from land size, scale factor and serial correlation) and shows its relationship with 'real' changes in prices (price of fertilizer relative to moving average prices of other inputs). Simple plots show both farm income and use of fertilizer increase from 1910 to 1945, then farm income plateaued but fertilizer consumption kept growing. Use of plant nutrients increased as real price paid for fertilizer decreased. Author tests the hypothesis that the substantial increase in the use of fertilizer has come about essentially as a response to a fall in the relative price of fertilizer; a fall both relative to the prices received for crops and to the prices paid for other inputs. The model consists of two parts: a long run demand function and an adjustment equation. The demand function considers fertilizer consumption to be a function of all the relevant product and input prices, permitting both output and other inputs to vary.
 
 $$ y_t=ba_0+ba_1x_2+ba_2x_2+(1-b)y_{t-1}+bu_t $$
 
 Where $y_t$ is log of fertilizer consumption (total plant nutrient use), $x_1$ log price of fertilizer relative to the prices received for crops, $x_2$ is the log price of fertilizer relative to the prices paid for other factors of production and $u_t$ is a disturbance. $b$ is the slope of adjustment equation
 
 $$y_t-y_{t-1}=b(y^*_t-y_{t-1})$$
 
 where $y^*_t$ is the log desired fertilizer consumption. Fertilizer is an weighted aggregate of nutrient factors from different nitrogen, phosphoric acid, and potash compounds ($N, P_2O_5$ and $K_2O$), assuming a perfect substitution. In addition to the basic model, author divides $y_t$ by the index of cropland used in some regressions, whereas $x_t$ is divided by prices received index at a point of time, and by prices received index with moving averages. Basically such divisions show the change in price of fertilizer relative to the other input prices. After controlling for serial correlation the elasticity of use of fertilizer with respect to its price becomes -0.529. Using the values from 1911-1933, the model does a nice work in predicting fertilizer demand for 1934-1956 period. That is, there may not have been any substantial structural change in the pattern of farmers' behavior with respect to fertilizer.

Griliches, Zvi. "The demand for inputs in agriculture and a derived supply elasticity." Journal of Farm Economics 41, no. 2 (1959): 309-322.

The study estimates demand functions for selected U.S. farm inputs (fertilizer, tractors, and hired labor), and shows possible relevance of these results for supply analysis. The model assumes that current plant nutrient (fertilizer) consumption is a function of the 'real' price of fertilizer (relative to the prices received for farm crops, prices of other inputs, and lagged plant nutrient consumption). The model follows the same framework as the study above and does not include farm income. It uses three dependent variables, $y_t$, total plant nutrients, total plant nutrients per acre of cropland used for crops, and price weighted plant nutrients used per acre of cropland used for crops. Indipendent variables, $x_t$, include price paid per plant nutrient unit divided by the index of prices received for products (all crops, specific crop, value of crop per acre). Price elasticity of fertlizier is about -0.5 in the short run and -2.0 in the long run. Regions (e.g., Southern regions) that use fertilizer for long time react faster to a price change. The author compared the price elasticity of total fertilizer used on cotton with the price elasticity of fertilizer used cotton per acre. The difference gives price elasticity of cotton acres i.e., the supply elasticity of cotton acres to be 0.3 in the short run and 0.5 in the long run. For instance, assume $Q$ is total fertilizer used on cotton, $F$ is the quaantity of fertilizer used per cotton acre and $A$ is the number of acres devoted to cotton.

$$Q=F.A\Rightarrow \frac{dQ}{dp}=\frac{dF}{dP}.A+\frac{dA}{dp}.F$$

$$\frac{dQ}{dp}\frac{p}{FA}=\frac{dF}{dp}\frac{p}{F}+\frac{dA}{dp}\frac{p}{A}$$

Thus, price elasticity of total fertilizer used on corron equates the price elasticity of fertilizer per cotton acre plus the price elasticity of cotton acres.
The short run 'real' price elasticity of the demand for tractor is -0.25, that of long run is -1.50. Short run elasticity of hired labor with respect to 'real' price is -0.10, and for the long run it is -0.44.
The author derives supply elasticity using

$$ e_{qp}=\sum_i k_i\mu_{ip} $$

Where, $e_{qp}$ is the supply elasticity of output $q$ with respect to its own price $p$, $k_i$ is the share of factor $i$ in total cost or revenue, $\mu_{ip}$ is the elasticity of demand for factor $i$ with respect to changes in the price of the product. The implied aggreagate supply elasticity is about 0.30 in the short run, and about 1.20 in the long run.
